### Autoregressive assumption
Autoregressive assumption는 모델이 아니라 언어를 다루는 방식(프레임워크)
$x=[x_1,x_2,\dots, x_n$] 일때, $p(x)$  
$p(x)=p(x_1)p(x_2\mid x_1)p(x_3 \mid x_2,x_1) \dots p(x_n \mid x_1, x_2, \dots, x_n-1)$  
$=\displaystyle\prod_{i=1}^{n} p(x_i \mid x_1, x_2, \dots, x_{n-1})$

n-gram을 이용해서 모델을 설계해야 함.

```python
# 간단한 corpus의 확룔 계산
# 리스트를 이용한 기본

text1 = "The cat is on the mat"
text2 = "The cat is on the floor"
text3 = "The dog is on the mat"

tokken1 = text1.split()
tokken2 = text2.split()
tokken3 = text3.split()

print(tokken1)
print(tokken2)
print(tokken3)
```
